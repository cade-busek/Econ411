# Growth versus Value Returns over the Last Economic Cycle

Since before the tech bubble burst, people in the investing world have been saying that value investing is dead. This has come from academics and professional investors, but probably most damning for me was how Warren Buffett, the most famous value investor in the world, has talked about what would be considered growth versus value stocks. He sums up his idea in his 1989 annual letter (without using the term growth investing) in this quote:
“It’s far better to buy a wonderful company at a fair price than a fair company at a wonderful price (Buffett).” 
This is a major endorsement of growth investing as opposed to value investing. On top of that, I later found out that the father of value investing, Benjamin Graham, made more money on one growth stock that he made on all of his other investments, combined. The stock was GEICO and it returned over 500x in 25 years, making him $400M and eventually his mentee Warren Buffet would buy the company out in its entirety (Khandelwal, 2022). All of this made me very excited at the start of this project to show that value investing was dead and had underperformed over the last economic cycle. We can say without question that many of the best performing stocks of the last twenty years have been growth stocks in technology such as Google, Apple and Amazon, and one thing that we know about these types of stocks is that they have never been statistically cheap by traditional measures. However, as I investigated the data regarding growth versus value stocks, three things became clear. First, traditionally cheap value stocks have not underperformed over the 13-year timeframe that I looked at; second this is trend of value performing well is likely to continue in the future. The last thing that was clear from the data was that while value can outperform, you will need to have the ability to pick Buffet’s wonderful companies at fair prices if you want the best performance.


When discussing value versus grow stocks, the first question that needs to be asked is what is a cheap stock? For the purposes of this discussion, we will use the metrics defined by Ron Bird and & Lorenzo Casavecchia in their paper Sentiment and financial health indicators for value and growth stocks: The European experience. They identified three main valuation metrics that are often used to determine if stocks are cheap or if they are expensive. Those three are book to market, sales to price, and earnings to price (Bird & Casavecchia, 2007). In this discussion we will use the inverse of those metric because they are more common.  In this paper, we will look at data from the last complete economic cycle which for this discussion will be defined as the low point for the S&P 500 during the housing crash, which was March 9th, 2009, until the low point of the S&P 500 in the current economic downturn which occurred on September 29th, 2022. The data shows that during this timeframe, value has performed at the same level as growth, and we use this to determine if value can continue to keep up with growth in the future. We will use the metrics defined above to show that unless you have the ability to successfully choose growth (or value) stocks extremely well, it does not matter if you buy a basket of value or growth stocks.

The purpose of trying to pick stocks of course is to try to beat the stock market generally. According to Louis K.C. Chan and Josef Lakonishok’s paper Value and Growth Investing: Review and Update, value tends to outperform growth stocks over long periods of time although others will argue that value is dead at this point due to its own long period of underperformance vs growth stocks(Chan & Lakonishok, 2004). In this paper, we show that value has not underperformed the market or growth stocks over the last economic cycle. For the market as a whole, we will use the S&P 500 as our benchmark that we want to see growth or value can beat. During that time, the S&P returned roughly 15% when you assume dividends were reinvested over this period. This means we need to try to find a strategy that will produce a return of better that 15% over that same period. It is important to note that these returns are before taxes. Taxes will be ignored for this paper except for when it is explicitly said that they are being taken into consideration. When leaving out taxes and just considering annual return, it is striking how similar the two styles of investing performed particularly when looking at high versus low price-to-earnings companies.

The data from this screener which excluded companies with negative earnings from the price-to-earnings group, which provided a very basic but imperfect screener for quality. By leaving out companies with negative price-to-earnings, some companies growing revenue rapidly might have been incorrectly left out of the analysis based on a price-to-earnings basis. Additionally, any value companies that had a bad quarter may have been excluded. For instance, bank stocks that struggled during the financial crisis may have been left out of this screener. The screener for price-to-earnings had the top and bottom 25th percentile of companies with over a billion-dollar market cap on March 9th, 2009.

```python
data = df.iloc[:, [3, 5,7, 8]]


data
```

```python
reg_vars = list(data.keys())

reg_vars
```

```python
y_var = [reg_vars[-2]]
x_vars = reg_vars[:-2] + reg_vars[-1:]

reg_data = data[reg_vars]
reg_data.corr().round(2)
```

It is interesting to not how little correlation there appears to be between price-to-earnings and returns over the period, although there does appear to be a slight tilt towards companies that are lower price-to-earnings.

```python
import statsmodels.api as sm
y = data['Total Return:20090309:20220929 (annualized including dividends)']
print(y_var)

x = data[["P/E:20090309", 'Total Return:20090309:20220929 (including dividends)', 'Total Returns in B Dollars']]



x["Constant"] = 1
sm.add_constant(1)
results = sm.OLS(y, x).fit()
results.summary()
```

Ultimately, we can see an R-squared of 0.590, which show that there is certainly predictability from the variables used, but it is extremely low and price to earnings is not alone enough to be predictable of stock returns.

In addition to statistical analysis, we can look at how would you do if you were to invest $1 in each of the 162 value companies and a dollar in each of the 162 growth companies. The value companies would have returned 16.37\% annually. On the other hand, investing one dollar in each high price-to-earnings growth stock resulted in a return of 16.42\% annually over the same period, which is essentially identical performance.

While value and growth stocks ended up with nearly identical returns, they as groups are getting there with different paths. The growth groups has more of the highest and lowest performing stocks while the value companies had fewer outliers on either the high end or low end of performance. The scatter plot and some data analytics shows that basically anyone could have done relatively well picking stocks over this time period as out of the 324 stocks that came through the price-to-earning screener, exactly two-thirds (216) produced at least of 10% annually with dividends reinvested, regardless of if these stocks were in the low or high price-to-earnings category.

While 10% is a good return, we are hoping for at least a 15% return to bear the S&P 500,  and in the screener 144 out of the 324 stocks in the screener produced returns of at least 15% annually over the timeframe. Companies with price-to-earnings of less than 6.75 (the lowest 25th percentile) had a 48% chance of returning at least 15% annually with 79 of the 162 companies beating that hurdle rate. For the companies with a price-to-earnings of over 13.75 (the highest 25th percentile) these companies had a 40.12% chance of achieving a return of at least 15% annual over the allotted period. However, as mentioned earlier, growth companies had significantly greater variation in their performance which is why you can perform far better if you can successfully pick growth companies. Take two theoretical companies, one producing 15% annually and one producing 25% annually. The company producing a 25% annual return provides over twice the total return of a company that produces our benchmark 15% annual returns after 10 years.  We will call the companies that produce at least 15% annual returns outperformers and companies with at least 25% annual returns significant outperformers. While the data shows that you are more likely to randomly pick an outperformer by choosing a value company, your odds of finding a significant outperformer are lower as the companies that are significantly outperformers are highly concentrated among the growth companies. If you just look for significant outperformers, growth companies were 50% more likely to have greater than a 25% annual return. Only seven of the 162 value companies were able to meet the 25% annual threshold whereas 11 of the growth companies were able to meet that threshold. If we move out bar up even a little higher to 28%, the number improve to 5 growth companies versus only one value company. We could keep going on and on, but the data shows very clearly the higher the return you set as your hurdle rate, the more likely it is that you will need to find a growth company to hit that hurdle rate. There are a couple of important limitations to the work that I did that do need to be remembered when looking at this data. First, when thinking about stock returns there are a couple of important things that we should consider when thinking about growth stock returns versus value stock returns and both items come down to the idea of how dividends could affect returns. First, let’s consider two companies that produce a 15% return each to shareholders, one pays out all of their earning while the other company retains them, are those returns guaranteed to produce the same returns for shareholders? Assuming that the companies’ values are growing in line with earnings, then the returns for the two companies will be very different. According to Fama and French in The Anatomy of Value and Growth Stock Returns, there are three ways for a stock to create returns. The first way is through dividends, the seconds is through an increase in increasing earnings by retaining earnings for reinvestment. The third way is for multiples expansion (higher P/E, P/B etc.) (Fama & French, 2019).  For the sake of this discussion, we will ignore multiples expansion and only focus on retention of earnings vs the paying out of dividends to see how they affect returns because multiple expansion will be random and not something that investors can project, so investors are unlikely to benefit from multiples expansion as a group. It turns out that retaining earnings or paying them out as dividends can make realized returns very different. Let’s for the sake of argument assume that the growth stock mentioned about pays out zero dividends while the value stock is paying out all of its earnings in dividends. After 10 years, a dollar invested in the growth stock has gone up 305% in value.  What about the dollar invested in a value stock? It has only gone up 150%, the dollar you originally invested plus 15% annual dividend that you received. You may be thinking that if you are able to reinvest those dividends back into the company you will do just as well, and you would be correct, if the stock price never moves (if the stock went up, you would underperform, if it went down, you would overperform). The other major assumption that this makes is that you are paying no taxes on the dividends you are receiving. If you figure in a 20% tax rate of dividends before you can reinvest, you wind up with a 211% gain after 10 years compared to a 244% gain after ten years. By having to pay taxes annually of dividends, you are giving up 9.5% of your total return over the 10-year period compared to if the company were able to reinvest their profits for you and you only having to pay taxes on the capital gains upon the sale of your stock. People may favor dividends for reasons that are not purely economical, but these reasons are beyond the scope of this discussion. Ultimately, this tax efficiency that growth investing has over value investing could be a reason to be more bullish on growth investing although taxes are never a good reason to make on investment decision.

While growth performed just as well as value over the last economic cycle, these seems unlikely to last in the future. There have been several things that have occurred over the last 15 years that have made growth companies perform better than their fundamentals might suggest they should. The first of these is having very low interest rates for a long period of time has pulled future returns forward and if returns were taken from the future, then returns must be given up in the future to compensate. Another reason that I have questions about growth companies as a whole going forward is the fact that companies are staying private longer. Companies staying private longer is causing more of the returns of the most highly sought-after growth companies to be made before the companies are available for the public to invest in. This is again causing the returns from growth companies to be pulled forward even more and in many cases, public market investors are being stuck with extremely overvalued companies such as Uber, which is down roughly 25% since it had its massive $82B IPO in May of 2019 (Merced & Conger, 2019). The last thing to remember is that the pandemic was very beneficial to companies that were growing quickly and had a lot of potential, even if they did not have the best track record. This again is another pull forward of future return that will probably have to be given back. Even worse in this case, many people assumed that trends from the pandemic such as e-commerce and working from home would continue at the same rate or continue growing after the pandemic ended. Unfortunately for companies who overinvested in this idea (and their investors), when these trends came down from pandemic highs, these stocks got hit hard and we have seen companies like Zoom and Peloton down significantly and they both may struggle to stay independent in the future. For these reasons, I think that value investing will continue to perform well into the future, and although growth companies are almost certain to have the highest individual performances, I think the chances of growth companies outperforming value as a group are significantly lower in the next decade than over the past decade. Ultimately though, returns for investors generally are probably not going to be as good going forward due to the low-rate environment that we talked about earlier as well as more and more smart people entering the investing field trying to take away any returns above the risk adjusted rate. These facts lead me to believe that most people, if they have the correct temperament, should use indexing to get the market returns and not worry if growth outperforms value or vice versa in the future.
