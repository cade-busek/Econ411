#### 2. Use the variance function from the textbook to find the variance of this list. Assume that the list represent a population in whole. \n",
    

```python
def variance(list1, sample = True):
    n = len(list1)
    list_mean = sum(list1)/n
   
    DoF = n-1
    sum_sq_diff = 0
    
    for val in list1:
        diff = val - list_mean
        sum_sq_diff += (val - list_mean)**2
#        print(val,list_mean**2, sum_sq_diff)
    if sample == False:
        variance_ = sum_sq_diff/n
    else:
        variance_ = sum_sq_diff/DoF
    return variance_
            
variance(list1, sample = False)
```
