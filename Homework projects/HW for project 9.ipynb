1. Rerun the last regression where, only including oecd countries in the subset. A list of OECD countries is provided. You will need to find the 3 letter ISO code associated with each country. (Hint: Reimport the dataframe we saved early in the lesson)

```python
import matplotlib.pyplot as plt
import pandas as pd
hw_data = pd.read_excel("EFWAndRGDP.xls", index_col = [0])
hw_data
```

```python
oecd_countries = ["Australia", "Austria", "Belgium", "Canada", "Chile", "Colombia", "Costa Rica",
            "Czech Republic", "Denmark", "Estonia", "Finland", "France", "Germany",
            "Greece", "Hungary", "Iceland", "Ireland", "Israel", "Italy", "Japan", 
            "Latvia", "Lithuania","Luxembourg", "Mexico", "Netherlands", "New Zealand", "Norway",
             "Poland","Portugal", "Slovakia", "Slovenia", "South Korea", "Spain", "Sweden",
             "Switzerland", "Turkey", "United Kingdom", "United States"]
```

```python
ahw_data = hw_data[hw_data["Countries"].isin(oecd_countries)]
ahw_data
```

```python
del hw_data['ISO_Code_2']
del hw_data['Year']
del hw_data['World Bank Region']
del hw_data['Countries']
del hw_data['World Bank Current Income Classification, 1990-present (L=Low income, LM=Lower middle income, UM=Upper middle income, H=High income)']
hw_data
```

```python
reg_vars = list(hw_data.keys())
reg_vars
```

```python
y_var = [reg_vars[-1]]
x_vars = reg_vars[:-1]
reg_data = hw_data[reg_vars]
reg_data.corr().round(2)
```

```python
import statsmodels.api as sm
r_df = reg_data.dropna(axis = 0, how = "any")
y = r_df[y_var]
X = r_df[x_vars]
X["Constant"] = 1
results = sm.OLS(y, X).fit()
r_df["Predictor"] = results.predict()
r_df['Residuals'] = results.resid
results.summary()
```

3. Plot seperate scatter plots for each variable (x-axis) against the predictor (y-axis). ThenPlot seperate scatter plots for each variable (x-axis) against the predictor (y-axis). Do we observe bias in the resdiuals when controlling for each variable? Why does this matter?



```python

def plot_residuals(r_df, y_var, x_vars):
    for x_var in x_vars:
        fig, ax = plt.subplots(figsize = (14,10))
        r_df.plot.scatter(x = x_var,
                         y = "Predictor", 
                          s = 30, ax = ax)
        plt.xticks(rotation=90)
        plt.show()
        plt.close()
    
        


plot_residuals(r_df, y_var, x_vars)
```

There is certainly bias is legal system and property rights as well as regulation and to a lesser extent in freedom to trade internationally. This shows the importance of these things to a strong economy.

```python
predictor = results.predict()
predictor_series = pd.Series(predictor)
ahw_data = ahw_data.reset_index(drop=True)
ahw_data[y_var[0] + 'Predictor'] = predictor_series
ahw_data = ahw_data.rename(columns={y_var[0] + 'Predictor': 'Predictor'})

ahw_data
```

4. Plot the observed and predicted values of RGDP Per Capita for each individual country.


```python
def plot_residuals(data):
    fig, ax = plt.subplots(figsize=(14,10))
    ax.scatter(x=data['RGDP Per Capita'],
               y=data['Predictor'],
               s=50)
    ax.set_title(data['Countries'].iloc[0])
    ax.set_xlabel('RGDP Per Capita')
    ax.set_ylabel('Predictor')
    plt.xticks(rotation=90)
    plt.show()
    plt.close()

for country, data in ahw_data.groupby('Countries'):
    plot_residuals(data)
```
